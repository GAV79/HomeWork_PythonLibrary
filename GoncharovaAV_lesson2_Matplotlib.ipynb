### Задание 2

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 10, 51)
f = np.cos(t)
label_dict = {'color': '#c20078', 'fontsize': 14, 'fontweight': 'normal', 'fontstyle': 'italic', 'fontfamily': 'Calibri'}
plt.plot(t, f, color='green')
plt.title('График f(t)', color='blue', fontsize='16', fontweight='bold')
plt.xlabel('значения t', fontdict=label_dict)
plt.ylabel('значения f', fontdict=label_dict)
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show
```

### Задание 3*

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
title_dict = {'color': '#c20078', 'fontsize': 14, 'fontweight': 'normal', 'fontstyle': 'italic', 'fontfamily': 'Calibri'}

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim(-5, 5)
ax1.set_title('график y1', fontdict=title_dict)

ax2.plot(x, y2)
ax2.set_title('график y2', fontdict=title_dict)

ax3.plot(x, y3)
ax3.set_title('график y3', fontdict=title_dict)

ax4.plot(x, y4)
ax4.set_title('график y4', fontdict=title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import pathlib # для определения рабочей директории
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
plt.style.use('fivethirtyeight')

work_path = pathlib.Path.cwd() # сохраняем путь до рабочей директории в переменную
data_path = Path(work_path, 'dataset', 'creditcard.csv')

cre = pd.read_csv(data_path)

print(f'Число наблюдений(строки) и признаков(столбцы): {cre.shape}')
print()
print(f'перечень признаков: {cre.columns}')
print()
cre_class_series = cre.value_counts("Class")
print('Число наблюдений для каждого значения признака "Class":')
print(cre_class_series)
print()
cre_class_series.plot(kind='bar', linewidth=2, edgecolor='#000000')
```

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import pathlib # для определения рабочей директории
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
plt.style.use('fivethirtyeight')
from pylab import rcParams

rcParams["figure.figsize"] = 9, 12
work_path = pathlib.Path.cwd() # сохраняем путь до рабочей директории в переменную
data_path = Path(work_path, 'dataset', 'creditcard.csv')

cre = pd.read_csv(data_path)
cre_Class0 = cre[cre['Class'] == 0]
cre_Class1 = cre[cre['Class'] == 1]
cre_V1_Class0 = cre_Class0['V1'].value_counts(bins=20, ascending=True)
cre_V1_Class1 = cre_Class1['V1'].value_counts(bins=20, ascending=True)
cre_V1_Class0.head(11)
cre_V1_Class0.plot(kind='bar', logy=True, alpha=0.5, color='grey', label='Class=0')
cre_V1_Class1.plot(kind='bar', logy=True, alpha=0.5, color='red', label='Class=1')
plt.xlabel('V1', color='blue', fontsize='16', fontweight='bold')
plt.legend(loc='upper left')
# не разобралась с аргументом density
```
