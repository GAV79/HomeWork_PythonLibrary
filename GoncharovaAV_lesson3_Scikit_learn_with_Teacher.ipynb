### Задание 1

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as pd
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

boston = load_boston()
print(boston.keys())
print()
data = boston['data']
feature_names = boston['feature_names']
target = boston['target']

X = pd.DataFrame(data, columns=feature_names)
print(X.head(4))
print()
X.info()
print()

y = pd.DataFrame(target, columns=['price'])
print(y.head(4))
print()
y.info()
print()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test.info()
print()

lr = LinearRegression()
lr.fit(X_train, y_train) # обучение по тренировочной выборке

y_pred = lr.predict(X_test) # на тестовой выборке получаем рассчитанные по построенной модели значения

check_test = pd.DataFrame({"y_test": y_test["price"], "y_pred": y_pred.flatten()})
print(check_test.head(10))
print()
w = lr.intercept_
print(f'Свободный коэффициент модели = {w}')
print()
w_coef = lr.coef_
print(f'Коэффициенты регрессионной модели = {w_coef}')
print()
print(f'Уравнение линейной регрессии: цена = '
      f'{round(w[0], 2)}{round(w_coef[0, 0], 3)}*{feature_names[0]}+'
      f'{round(w_coef[0, 1], 3)}*{feature_names[1]}+'
      f'{round(w_coef[0, 2], 3)}*{feature_names[2]}+'
      f'{round(w_coef[0, 3], 3)}*{feature_names[3]}'
      f'{round(w_coef[0, 4], 3)}*{feature_names[4]}+'
      f'{round(w_coef[0, 5], 3)}*{feature_names[5]}'
      f'{round(w_coef[0, 6], 3)}*{feature_names[6]}'
      f'{round(w_coef[0, 7], 3)}*{feature_names[7]}+'
      f'{round(w_coef[0, 8], 3)}*{feature_names[8]}'
      f'{round(w_coef[0, 9], 3)}*{feature_names[9]}'
      f'{round(w_coef[0, 10], 3)}*{feature_names[10]}+'
      f'{round(w_coef[0, 11], 3)}*{feature_names[11]}'
      f'{round(w_coef[0, 12], 3)}*{feature_names[12]}'
     )

print()
R2 = r2_score(y_test, y_pred)
print(f'Коэффициент множественной детерминации (R2) = {round(R2, 2)}')
```

## Задание 2

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as pd
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

boston = load_boston()
data = boston['data']
feature_names = boston['feature_names']
target = boston['target']
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=['price'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
model.fit(X_train, y_train.values[:, 0])
y_pred = model.predict(X_test)

check_test = pd.DataFrame({"y_test": y_test["price"], "y_pred": y_pred.flatten()})
print(check_test.head(10))
print()
R2 = r2_score(y_test, y_pred)
print(f'Коэффициент множественной детерминации (R2) = {round(R2, 2)}')
print()
print('Так как, чем ближе R^2 к 1, тем выше качество модели, то модель построенная на алгоритмe «случайный лес» работает лучше')
```

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine

wine = load_wine()
print('ключи датасета:')
wine_keys = [wine.keys()]
print(wine_keys)
print()
print('количество строк и столбцов')
data = wine['data']
print(data.shape)
print()
print('Названия признаков:')
feature_names = wine["feature_names"]
print(feature_names)
print()
print('Описание датасета:')
print(wine['DESCR'])
print()
print('Количество и названия классов целевой переменной:')
target_names = wine['target_names']
print(len(target_names))
print(target_names)
print()
print('Создаем датафрейм Х:')
X = pd.DataFrame(data, columns=feature_names)
print(X)
print()
print('количество строк и столбцов датафрейма Х:')
print(X.shape)
print()
print('Пропущенные значения датафрейма Х:')
print(X.info())
print('Пропущенных значений нет')
print()
print(X.columns)
print()
print('добавляем столбец target')
X['target'] = wine.target
X = X.astype({'target':np.int64})
print(X.columns)
print()
print(f'Тип данных "target": {X["target"].dtype}')
print()
X_corr = X.corr()
X_corr = np.round(X_corr, 2)
X_corr[np.abs(X_corr) < 0.3] = 0
print('Корреляционная матрица:')
print(X_corr)
print()
print('Коэффициент корреляции с целевой переменной > 0.5')
high_corr = X_corr['target']
high_corr = high_corr[(np.abs(high_corr) > 0.5) & ~(high_corr == 1)]
print(high_corr)
print()
print('Удаляем "target" из датафрейма Х:')
X.drop('target', axis=1, inplace=True)
print(X.columns)
print()
print('Добавляем столбцы с квадратами значений переменных с коэфф.корр > 0,5')
a = np.array(high_corr.index)
for i in a:
    X[i +'_2'] = X[i] * X[i]
print(X.columns)
print()
X.describe()
```
